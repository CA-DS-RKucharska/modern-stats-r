## **exercise 2.6:** means and correlations

```python
%%R
# 1. means of each variable
mean_height <- mean(measurements$height); mean_weight <- mean(measurements$weight)
mean_measurements <- data.frame(mean_height,mean_weight); mean_measurements
```

```python
%%R
# 2. pearson and spearman correlations between the two variables
pearson <- cor(measurements$height,measurements$weight); spearman <- cor(measurements$height,measurements$weight,method='spearman')
correlations <- data.frame(pearson, spearman); correlations
```

## **exercise 2.14:** plotting log-transformed values

## **exercise 2.15:** facet wrapping

```python
%%R
# 1. facet by cut
ggplot(diamonds, aes(carat, price)) +
  facet_wrap(~ cut) +
  xlab("Diamond weight (carat)") +
  ylab("Price (USD)") +
  geom_point(alpha=0.25)
```

```python
%%R
# 2. change the rows
# ?facet_wrap
ggplot(diamonds, aes(carat, price)) +
  facet_wrap(~ cut, nrow = 5) +
  xlab("Diamond weight (carat)") +
  ylab("Price (USD)") +
  geom_point(alpha=0.25)
```

## **exercise 2.16:** BOXPLOTS !!!!

```python
%%R
# 1. prices by cut
ggplot(diamonds, aes(cut, price)) +
  geom_boxplot() +
  xlab("Diamond cut") +
  ylab("Price (USD)")
```

```python
%%R
# 2. changing colours
# looked on here: https://r-graph-gallery.com/264-control-ggplot2-boxplot-colors.html
ggplot(diamonds, aes(cut, price)) +
  geom_boxplot(fill='orange',color='red') +
  xlab("Diamond cut") +
  ylab("Price (USD)")
```

```python
%%R
# 3. reorder cut by the median price of each bucket
# 4. add jitter
ggplot(diamonds, aes(reorder(cut, price, median), price)) +
  geom_boxplot(fill='orange') +
  xlab("Diamond cut") +
  ylab("Price (USD)") +
  geom_jitter(size = 0.1, alpha = 0.2)
```

# **exercise 2.17:** histograms :/

```python
%%R
# 1. hist of diamond prices
hist(diamonds$price)
```

```python
%%R
# 2. hist of prices faceted by cut
# 3. use col = I('color') to add an outline (https://www.tutorialspoint.com/change-the-outline-color-for-histogram-bars-using-ggplot2-in-r)
ggplot(diamonds, aes(price)) +
  facet_wrap(~ cut) +
  geom_histogram(col = I('black')) +
  xlab("Price (USD)")
```

# 8. plotting categorical data
## **exercise 2.18**: bar charts :>

```python
%%R
# 1. diamonds by cut
barplot(table(diamonds$cut))
```

```python
%%R
# 2. colours!
# 3. widths!
ggplot(diamonds, aes(cut, fill=cut)) +
  geom_bar(width = 0.5) +
  xlab("Diamond cut") +
  ylab("Frequency")
```

```python
%%R
# 4. fill with clarity
# 5. get dodgy
ggplot(diamonds, aes(cut, fill=clarity)) +
  geom_bar(width = 0.5, position = "dodge") +
  xlab("Diamond cut") +
  ylab("Frequency")
```

```python
%%R
# 6. flip the axes
ggplot(diamonds, aes(cut, fill=cut)) +
  geom_bar(width = 0.5) +
  xlab("Diamond cut") +
  ylab("Frequency") +
  coord_flip()
```
